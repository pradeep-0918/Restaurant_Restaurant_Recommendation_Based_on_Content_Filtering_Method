Task 2: ***Restaurant Restaurant Recommendation Based on Content Filtering Method***

```python
import pandas as pd
import numpy as np
df = pd.read_csv("/content/Dataset .csv")
print("Dataset is loaded Successfully")
```

```python
print("Handling Missing Values")
df.isnull().sum()
```

```python
print("Fill missing Value using by taking mode()")
df['Cuisines'] = df['Cuisines'].fillna(df['Cuisines'].mode()[0])
df.isnull().sum()
```

```python
columns_needed = ['Restaurant Name', 'Cuisines', 'Average Cost for two', 'Aggregate rating', 'City', 'Country Code']
df = df[columns_needed]
```

```python
print("Encoding Categorical into Numerical")
df_encoded = pd.get_dummies(df, columns=['Cuisines', 'City'])
df_encoded
```

```python
print("Processed Dataset:")
print(df_encoded.head())
```

```python
print("Defining User Preferences.")
user_cuisine = input("Enter your preferred cuisine (e.g., Italian, North Indian): ").strip()
user_budget = float(input("Enter your maximum budget for two (e.g., 500): "))
user_location = input("Enter preferred location (optional - press Enter to skip): ").strip()
```

```python
print("Creates a blank DataFrame with same columns and adding '0' to the dataframe")
user_pref = df_encoded.iloc[0:0].copy()
user_pref.loc[0] = 0
```

```python
print("Checking given user_Cuisine is present in dataset/not")
cuisine_col = f'Cuisines_{user_cuisine}'
if cuisine_col in user_pref.columns:
    user_pref[cuisine_col] = 1
else:
    print(f" Cuisine '{user_cuisine}' not found in dataset.")
```

```python
print("Checking given user_location is present in dataset/not")
if user_location:
    location_col = f'Location_{user_location}'
    if location_col in user_pref.columns:
        user_pref[location_col] = 1
    else:
        print(f" Location '{user_location}' not found in dataset.")
```

```python
user_pref['Average Cost for two'] = user_budget
```

```python
print("similar the restaurant is to the user's preferences")
from sklearn.metrics.pairwise import cosine_similarity
restaurant_names = df['Restaurant Name']
df_features = df_encoded.drop(['Restaurant Name', 'Aggregate rating'], axis=1)
user_vector = user_pref[df_features.columns]

similarities = cosine_similarity(df_features, user_vector)
df['Similarity Score'] = similarities
```

```python
top_matches = df.sort_values(by='Similarity Score', ascending=False)
```

```python
print("\n Top 5 Recommended Restaurants:")
print(top_matches[['Restaurant Name', 'Cuisines', 'Average Cost for two', 'Aggregate rating', 'City', 'Similarity Score']].head(5))
```
